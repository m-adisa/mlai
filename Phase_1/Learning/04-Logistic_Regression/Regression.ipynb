```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Loss function: Log likelihood (cross-entropy loss)
def compute_loss(y, hx):
    return -np.mean(y * np.log(hx) + (1 - y) * np.log(1 - hx))
```

```python
# Gradient of the loss function
def compute_gradient(X, y, hx):
    return np.dot(X.T, (hx - y)) / y.shape[0]
```

```python
# Dataset
X = np.array([
    [2, 1],
    [3, 2],
    [4, 4], # Positive class
    [1, 5],
    [2, 6],
    [3, 7]  # Negative class
])

y = np.array([1, 1, 1, 0, 0, 0])  # Labels

# Plot the initial data
plt.scatter(X[:3, 0], X[:3, 1], color='blue', marker='o', label='Positive class')
plt.scatter(X[3:, 0], X[3:, 1], color='red', marker='x', label='Negative class')
plt.xlabel('Whisker Length')
plt.ylabel('Ear Flappiness Index')
plt.legend()
plt.title('Initial Data Distribution')
plt.show()
```

```python
# Plotting decision boundary function
def plot_decision_boundary(X, y, weights, loss, iteration):
    plt.scatter(X[:3, 1], X[:3, 2], color='blue', marker='o', label='Positive class' if iteration == 0 else "")
    plt.scatter(X[3:, 1], X[3:, 2], color='red', marker='x', label='Negative class' if iteration == 0 else "")

    # Extend the x values a bit beyond the minimum and maximum values of the dataset for better visualization
    x_values = np.array([np.min(X[:, 1]) - 1, np.max(X[:, 1]) + 1])
    y_values = -(weights[0] + weights[1] * x_values) / weights[2]
    plt.plot(x_values, y_values, "k")

    plt.xlim(x_values[0], x_values[1])
    plt.ylim(np.min(X[:, 2]) - 1, np.max(X[:, 2]) + 1)
    plt.xlabel('Whisker Length')
    plt.ylabel('Ear Flappiness Index')
    plt.title(f'Decision Bountary at Iteration {iteration} with Loss: {loss:.4f}')
    if iteration == 0:
        plt.legend()
    plt.show()
```

```python
# Logistic Regression Implementation
def logistic_regression(X, y, learning_rate=0.01, num_iterations=100): # epochs == num_iterations
    # Add intercept term to X (Augment feature matrix with a column of ones)
    X = np.hstack((np.ones((X.shape[0], 1)), X))

    # Initialize weights
    weights = np.zeros(X.shape[1])

    # Gradient Descent
    for i in range(num_iterations):
        # Compute linear combination
        z = np.dot(X, weights)

        # Apply sigmoid function
        hx = sigmoid(z)

        # Compute loss
        loss = compute_loss(y, hx)

        # Compute gradient
        gradient = compute_gradient(X, y, hx)

        # Update weights
        weights -= learning_rate * gradient

        # Plot decision boundary
        if i % 10 == 0 or i == num_iterations - 1:
            plot_decision_boundary(X, y, weights, loss, i)
            print(f"Iteration {i}: Theta = {weights}, Loss = {loss}")

    return weights
```

```python
# Train the model
final_weights = logistic_regression(X, y)
```
